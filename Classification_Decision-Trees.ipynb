```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
# Imagine that you are a medical researcher compiling data for a study. You have collected data about a set of patients, all of
# whom suffered from the same illness. During their course of treatment, each patient responded to one of 5 medications, Drug A,
# Drug B, Drug c, Drug x and y. 

# Part of your job is to build a model to find out which drug might be appropriate for a future patient with the same illness. 
# The feature sets of this dataset are Age, Sex, Blood Pressure, and Cholesterol of patients, and the target is the drug that
# each patient responded to. 

df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/drug200.csv')
```

```python
df
```

```python
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
```

```python
from sklearn import preprocessing

le_sex = preprocessing.LabelEncoder()
le_sex.fit(['F', 'M'])

X[:, 1] = le_sex.transform(X[:, 1])

le_BP = preprocessing.LabelEncoder()
le_BP.fit(['LOW', 'NORMAL', 'HIGH'])
X[:, 2] = le_BP.transform(X[:, 2])

le_Chol = preprocessing.LabelEncoder()
le_Chol.fit([ 'NORMAL', 'HIGH'])
X[:,3] = le_Chol.transform(X[:,3])
```

```python
y = df['Drug']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
dugtree = DecisionTreeClassifier(criterion='entropy', max_depth=4)
dugtree
```

```python
dugtree.fit(X_trainset, y_trainset)
```

```python
predtree = dugtree.predict(X_testset)
```

```python
from sklearn.metrics import jaccard_similarity_score
```

```python
print(jaccard_similarity_score(predtree, y_testset))
```

```python
print (predtree[0:5])
print (y_testset[0:5])
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_testset, predtree))
```

```python
# accuracy without sklearn

np.sum(predtree==y_testset)/len(predtree)
```
